```python
import numpy as np
```

```python
file  = open('input/input_day04.txt','r')
file_content = file.readlines()
```

```python
separated_content = file_content
for i in range(0,len(file_content)):
    separated_content[i] =  list(((separated_content[i]).replace("\n", "")))

```

```python
matrix_content = np.array(separated_content)
```

```python
xmas = 0
a = 139
for x in range(0, matrix_content.shape[0]):
    for y in range(0, matrix_content.shape[1]):
        if matrix_content[x, y] == "X":
            if y-3 >= 0 :
                if(([matrix_content[x, y-1], matrix_content[x, y-2], matrix_content[x, y-3]] == ["M", "A", "S"])):
                    xmas += 1
            if y+3 <= a:
                if ([matrix_content[x, y+1], matrix_content[x, y+2], matrix_content[x, y+3]] == ["M", "A", "S"]):
                    xmas += 1
            if x-3 >= 0:
                if ([matrix_content[x-1, y], matrix_content[x-2, y], matrix_content[x-3, y]] == ["M", "A", "S"]):
                    xmas += 1
            if x+3 <= a:
                if ([matrix_content[x+1, y], matrix_content[x+2, y], matrix_content[x+3, y]]  == ["M", "A", "S"]):
                    xmas += 1
            if (x-3 >= 0) and (y-3 >= 0):
                if ([matrix_content[x-1, y-1], matrix_content[x-2, y-2], matrix_content[x-3, y-3]] == ["M", "A", "S"]):
                    xmas += 1
            if (x+3 <= a) and (y-3 >= 0):
                if ([matrix_content[x+1, y-1], matrix_content[x+2, y-2], matrix_content[x+3, y-3]] == ["M", "A", "S"]):
                    xmas += 1
            if (x-3 >= 0) and (y+3 <= a):
                if ([matrix_content[x-1, y+1], matrix_content[x-2, y+2], matrix_content[x-3, y+3]] == ["M", "A", "S"]):
                    xmas += 1
            if (x+3 <= a) and (y+3 <= a):
                if ([matrix_content[x+1, y+1], matrix_content[x+2, y+2], matrix_content[x+3, y+3]] == ["M", "A", "S"]):
                    xmas += 1
```
