```python
separated_content = file_content
instructions = []
code_input = []
for i in range(0,len(file_content)):
    if i/100 == round(i/100, 0):
        print(i, file_content[i])
        print(type(file_content[i]))
    
    if "|" in file_content[i]:
        separated_content[i] =  separated_content[i].replace("\n", "").split(sep="|")
        instructions = instructions + [separated_content[i]]

    elif "," in file_content[i]:
        separated_content[i] =  separated_content[i].replace("\n", "").split(sep=",")
        
        code_input = code_input + [separated_content[i]]

```

```python
instructions
```

```python
instructions_array = np.array(instructions)
```

```python
# solution
```

```python
instruction_dict = {}
for n in range(int(min(instructions_array[:,0])), int(max(instructions_array[:,0])) + 1):
    instruction_dict[str(n)] = []
    for i in range(0, len(instructions_array)):
        if instructions_array[i,0] == str(n):
            instruction_dict[str(n)] = instruction_dict[str(n)] + [instructions_array[i,1]]
```

```python
instruction_dict
```

```python
results_dict = {}
```

```python
l = 0
err_found = False

for line in code_input:
    print(l, "of", len(code_input))
    results_dict["line" + str(l)] = 1
    err_found = False
    for x in range(1, len(line)):
        # print(line[x])
        # print("checkpoint1")
        if ~err_found:
            # print("checkpoint2")
            for pre_x in range(0, x):
                # print(pre_x)
                # print("checking", line[pre_x])
                if line[pre_x] in instruction_dict[line[x]]:
                    err_found = True
                    print("err_found")
                    results_dict["line" + str(l)] = 0
    l += 1
```

```python
ans = []
for i in range(0, len(results_dict)):
    nm = "line"+str(i)
    if results_dict[nm] == 1:
        mid_num_index = int((len(code_input[i]) + 1) / 2) - 1 
        ans = ans + [int(code_input[i][mid_num_index])]
        print(code_input[i])

```
